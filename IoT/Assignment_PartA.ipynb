### Justification

The chosen data model is a reference data model. The reason reference data model is chosen because reference data model has better scalability compared to embedded data model. This is because related data are stored in separate collections and referenced using identifiers or references. Thus, reference data model is suitable for large data which is frequently updated. Besides, reference data model maintains data consistencies. Reference data model avoids data duplication by maintaining a single copy of related data. On the other hand, there is a risk of data duplication for embedded data model because related data are stored within a single document. Furthermore, reference data model has great flexibility when dealing with complicated relationships and evolving data structures. In our case, date is used as referencing for our reference data model.

```python
import pymongo
import csv
import json
from pprint import pprint
from pymongo import MongoClient

#Connect to a MongoDB server, select a database, and drop the selected database.
client = MongoClient('192.168.0.129', 27017) 
db = client.fit3182_assignment_db
client.drop_database('fit3182_assignment_db')
```

```python
#Establishes a connection to a specific database named "fit3182_assignment_db"
db = client.fit3182_assignment_db
```

```python
#Accesses the historic collection within the selected MongoDB database db
collection = db.historic
```

```python
import csv
import json

#The climate_collection and hotspot_collection lists will contain the data from the respective CSV files, 
#where each row is represented as a dictionary.
climate_collection = []
with open('Data/climate_historic.csv', 'r') as file1:
    reader = csv.DictReader(file1)
    for row in reader:
        climate_collection.append(row)

hotspot_collection = []
with open('Data/hotspot_historic.csv', 'r') as file2:
    reader = csv.DictReader(file2)
    for row in reader:
        hotspot_collection.append(row)
```

```python
#Documents from both the climate and hotspot collections are inserted into the MongoDB collection based on the 
#reference data model where corresponding hotspot data documents are inserted if they have the same date.
for doc1 in climate_collection:
    reference_data_model_doc1 = {
                "date": doc1['date'],
                "station": int(doc1['station']),
                "air_temperature_celcius": int(doc1['air_temperature_celcius']),
                "relative_humidity": float(doc1["relative_humidity"]),
                "windspeed_knots": float(doc1["windspeed_knots"]),
                "max_wind_speed": float(doc1["max_wind_speed"]),
                "precipitation": doc1['precipitation '], 
                "GHI_w/m2": int(doc1["GHI_w/m2"])
            }
    collection.insert_one(reference_data_model_doc1)
    for doc2 in hotspot_collection:
        if doc2['date'] == doc1['date']:
            reference_data_model_doc2 = {
                "child_date": doc1['date'],
                "latitude": float(doc2['latitude']),
                "longitude": float(doc2['longitude']),
                "datetime": doc2['datetime'],
                "confidence": int(doc2['confidence']),
                "surface_temperature_celcius": int(doc2['surface_temperature_celcius'])
            }
            collection.insert_one(reference_data_model_doc2)
   
            
```

```python
from pprint import pprint

#Outputs a formatted and easy-to-read representation of the documents' content.
for document in collection.find({}):
    pprint(document)
    
```

```python
#Find climate data on 12th December 2022
query = {"date": "12/12/2022", "latitude": {"$exists": False}}

results = collection.find(query)

for doc in results:
    print(doc)
```

#### b.

```python
#Find and show only the latitude, longitude, surface temperature (C), and confidence when the surface temperature (C) 
#was between 65 C and 100 C, and not display _id
query = {
    "surface_temperature_celcius": {
        "$gte": 65,
        "$lte": 100
    }
}

projection = {
    "_id":0,
    "latitude": 1,
    "longitude": 1,
    "surface_temperature_celcius": 1,
    "confidence": 1
}

results = collection.find(query,projection)

for doc in results:
    print(doc)
```

#### c.

```python
#Find date, surface temperature (C), air temperature (C), relative humidity and 
#max wind speed on 15th and 16th of December 2022, and not display _id
query = [
    {
        "$lookup": {
            "from": "historic",
            "localField": "date",
            "foreignField": "child_date",
            "as": "joined_data"
        }
    },
    {
        "$match": {
            "$or": [
                {"date": "15/12/2022"},
                {"date": "16/12/2022"}
            ]
        }
    },
    {
        "$unwind": "$joined_data"
    },
    {
        "$project": {
            "_id": 0,
            "date": 1,
            "surface_temperature_celcius": "$joined_data.surface_temperature_celcius",
            "air_temperature_celcius": "$air_temperature_celcius",
            "relative_humidity": "$relative_humidity",
            "max_wind_speed": "$max_wind_speed"
        }
    }
]
results = collection.aggregate(query)

for doc in results:
    print(doc)
```

#### d.

```python
#Display datetime, air temperature (C), surface temperature (C) and confidence when the confidence is between 80 and 100. 
query = [
    {
        "$lookup": {
            "from": "historic",  
            "localField": "child_date", 
            "foreignField": "date",
            "as": "joined_data"
        }
    },
    {
        "$match": {
            "confidence": {
                "$gte": 80,
                "$lte": 100
            }
        }
    },
    {
        "$unwind": "$joined_data"
    },
    {
        "$project": {
            "_id": 0,
            "datetime": "$datetime",
            "air_temperature_celcius": "$joined_data.air_temperature_celcius",
            "surface_temperature": "$joined_data.relative_humidity",
            "confidence": "$confidence"
        }
    }
]

results = collection.aggregate(query)

# Process the results
for doc in results:
    print(doc)
```

#### e.

```python
#Display top 10 records with the highest surface temperature (C).
query = [
    {
        "$sort": {
            "surface_temperature_celcius": -1
        }
    },
    {
        "$limit": 10
    }
]

results = collection.aggregate(query)

for doc in results:
    print(doc)
```

#### f.

```python
#Display the total number of fires and the relevant date
query = [
    {
        "$group": {
            "_id": "$child_date",
            "total number of fires": {
                "$sum": {
                    "$cond": [
                    { "$eq": [ { "$type": "$station" }, "missing" ] },
                    1,
                    0
                ]
                }
            }
        }
    },
  
    {
        "$project": {
            "_id": 0,
            "date": "$_id",
            "total number of fires": 1
        }
    }
]

results = collection.aggregate(query)

for doc in results:
    pprint(doc)
```

#### g.

```python
#Display the records of fires where the confidence is below 70
query = [
    {
        "$match": {
            "confidence": { "$lt": 70 }
        }
    }
]

results = collection.aggregate(query)

for doc in results:
    print(doc)
```

#### h.

```python
#Display average surface temperature (C) and the date
query = [
    {
        "$group": {
            "_id": "$child_date",
            "average surface temperature": { "$avg": "$surface_temperature_celcius" }
        }
    },
    
    {
        "$project": {
            "_id": 0,
            "date": "$_id",
            "average surface temperature": 1
        }
    }
]

results = collection.aggregate(query)

for doc in results:
    pprint(doc)
```

#### i.

```python
#Display the top 10 records with the lowest GHI
query = [
    {
        "$match": {
            "GHI_w/m2": { "$exists": True }
        }
    },
    {
        "$sort": {
            "GHI_w/m2": 1
        }
    },
    {
        "$limit": 10
    }
]

results = collection.aggregate(query)

for doc in results:
    print(doc)
```

#### j.

```python
#Display the records with a 24-hour precipitation recorded between 0.20 to 0.35
query = [
    {
        "$match": {
            "precipitation": { "$regex": '^ (0\.2[0-9]|0\.3[0-5])G$'}
        }
    }
]

results = collection.aggregate(query)

for doc in results:
    print(doc)
```

### Justification

The above reference data model design utilizes compound indexes. The reason compound indexes is used for the design of the reference data model is because the nature of the index having multiple fields together can help database to leverage the index for queries that match the fields in the index, leading to faster query execution. Furthermore, compared to single indexes, compound indexes can handle a wider variety of query patterns, thus helps to reduce the overall index storage overhead and optimize mongodb. In conclusion, the index design for the reference data model helps perform efficient queries based on multiple fields on complex relationships.
